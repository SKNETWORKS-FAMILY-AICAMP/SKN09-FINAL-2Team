```python
from new_chat_v2 import chat
# ✅ 예시 7: 이상한 질문들 모음
recipient_info = {
    'GENDER': "선택 안함",
    'AGE_GROUP': "10대 이하",
    'RELATION': "학교·멘토",
    'ANNIVERSARY': "합격·응원",
}

# user_inputs = [
#     "향초랑 디퓨저 차이가 뭐야?",
#     "느좋 분위기 선물 찾고 있음",
#     "센픽 브랜드 제품으로 찾아줘",
#     "exit"
# ]
user_inputs = [
    # 정상 질문
    "새로 들어온 초등학교 제자들에게 격려를 할 수 있는 선물을 주고 싶어",
    "귀여운 스타일로 부탁하고, 가격은 저렴한 걸로 해줘",

    # ❌ 선물과 무관한 질문
    "요즘 날씨 왜 이렇게 더워?",

    # ❌ 감정은 있지만 선물 문맥 아님
    "나 요즘 기분이 별로야",

    # ❌ 존재하지 않는 브랜드
    "디지몬코리아 브랜드 제품 보여줘",

    # ❌ 오타 + 추상 표현
    "쌘프란 선물 보여죠",  # (센스 + 프랑스?)

    # ❌ 신조어, 은어
    "그냥 꾸안꾸 스타일로다가 쓱 추천해줘",

    # ❌ 모호한 형용사
    "아늑하고 몽글한 선물 뭐가 있어?",

    # ❌ 상품명이면서 명확한 목적 없음
    "갤럭시",  

    # 🔵👌 브랜드 이름 + 형용사
    "구찌 느낌나는 선물로 찾아보려고 해."
]


chat(inputs=user_inputs, recipient_info=recipient_info)
```
